## Exercise 1: European Call

The value of a call option for a non-dividend-paying underlying stock in terms of the Blackâ€“Scholes parameters is:
\begin{align}
  C(S_t, t) &= N(d_1)S_t - N(d_2) PV(K) \\
     d_1 &= \frac{1}{\sigma\sqrt{T - t}}\left[\ln\left(\frac{S_t}{K}\right) + \left(r + \frac{\sigma^2}{2}\right)(T - t)\right] \\
     d_2 &= d_1 - \sigma\sqrt{T - t} \\
PV(K) &=Ke^{-r(T - t)}
\end{align}

The Notation (of the Parameters) are as follows: 

- $N(\cdot)$ is the cumulative distribution function of the standard normal distribution
- $T - t$ is the time to maturity (expressed in years)
- $S_t$ is the spot price of the underlying asset
- $K$ is the strike price
- $r$ is the risk free rate (annual rate, expressed in terms of continuous compounding)
- $\sigma$ is the volatility (finance)|volatility of returns of the underlying asset




```python
# 1: Black Scholes formula for European call

def bs_euro_call(St, t): 
    rt = T - t  # remaining time to maturity
    d1 = 1/ (sigma * rt ** 0.5) * (np.log(St/K) + (r + sigma ** 2 / 2) * rt)
    d2 = d1 - sigma * rt ** 0.5 
    PV_K = K * np.exp(-r * rt)
    
    C_St_t = norm.cdf(d1) * St - norm.cdf(d2) * PV_K
    
    return C_St_t
```

```python
r = 0.1 
sigma = 0.2 
X0 = 100
T = 1 
K = 100
C_t0 = bs_euro_call(X0, 0)

print ('The European call with S = ' + str(X0), ' at time t = ' + str(0), 'is equal to: ' + str(C_t0))
```

```python
# Compute the European call  via Monte Carlo Method Based on Black Scholes model
def sim_BS_Xt(T, nr_steps=1000, X0=100, r=0.1, sigma=0.2, nr_paths=1000): 
    
    step = T / nr_steps  # time grid: tg[0] = 0,  tg[-1] = T, tg length: n + 1 (t0, t1, ..., tn)
    dWs = 0 + step ** 0.5 * np.random.randn(nr_paths, nr_steps)  # dW: increments, follow a normal distribution
    exp_part = (r - sigma ** 2 / 2) * step + sigma * dWs 
    
    Xs = np.zeros((nr_paths, nr_steps + 1))  # each row is a path
    Xs[:,0] = X0 
    
    for i in range(nr_steps): 
        Xs[:,i+1] = Xs[:,i] * np.exp(exp_part[:,i]) 
        
    return Xs


```

```python
# simualate Xt by BS models
nr_steps=1000
nr_paths = 2000

step = T / nr_steps 
tg = np.arange(0, T + step, step)  # t0, t1, ..., tn (n+1) data point

Xs = sim_BS_Xt(T, nr_steps, X0, r, sigma, nr_paths)  # nr_paths * nr_steps (nr_path * (n+1))
plt.plot(tg, Xs.T)
plt.xlabel('time')
plt.ylabel('Xt')
plt.title('Simulated ' + str(nr_paths) + ' paths of Xt based on Black Scholes model')
plt.show()
```

```python
# price European call based on the simualated path of Xt (simulation according to Black Scholes model)
bs_euro_call_price = np.exp(-r * T) * np.mean(np.maximum(Xs[:,-1] - K,0))
print ('The European call based on BS simulation with S = ' + str(X0), 
       ' at time t = ' + str(0), 'is equal to: ' + str(bs_euro_call_price))
```

### Confidence Interval
A single value is meaningless if we do not give a confidence interval. Now we calculate the confidence interval.

According to central limit theorem:  
Let $\{X_1,..., X_n\}$ be a random sample of size $n$ â€” that is, a sequence of independent and identically distributed (i.i.d.) random variables drawn from a distribution of expected value given by $\mu$ and finite variance given by $\sigma$. Suppose we are interested in the ample mean|sample average: 

$$S_n := \frac{X_1+\cdots+X_n}{n}$$

Then, we have: 
        
$$\sqrt{n}\left(S_n - \mu\right)\ \xrightarrow{d}\ N\left(0,\sigma^2\right)$$   

The population variance $\sigma^2$ could be estimated by the sample variance $\sigma_n^2$


The sample variance is calculated as follows: 
$$s^2 = \frac{n}{n-1} \sigma_Y^2 = \frac{n}{n-1} \left( \frac{1}{n} \sum_{i=1}^n \left(Y_i - \overline{Y} \right)^2 \right) = \frac{1}{n-1} \sum_{i=1}^n \left(Y_i - \overline{Y} \right)^2 $$


Or, written as : 
$$\begin{align}
\operatorname{Var}(X) &= \operatorname{E}\left[(X - \operatorname{E}[X])^2\right] \\[4pt]
&= \operatorname{E}\left[X^2 - 2X\operatorname{E}[X] + \operatorname{E}[X]^2\right] \\[4pt]
&= \operatorname{E}\left[X^2\right] - 2\operatorname{E}[X]\operatorname{E}[X] + \operatorname{E}[X]^2 \\[4pt]
&= \operatorname{E}\left[X^2 \right] - \operatorname{E}[X]^2
\end{align}$$

```python
var_n = nr_paths /(nr_paths - 1) \
    * (np.mean(np.maximum(Xs[:,-1] - K,0) ** 2) - np.mean(np.maximum(Xs[:,-1] - K,0)) **2) 
    
std_n = var_n ** 0.5

# 95% interval corresponds to 1.96 * std_n / (sqrt(n))
half_int = 1.96 * std_n / nr_paths ** 0.5 
conf_interval = np.round([bs_euro_call_price - half_int,  bs_euro_call_price + half_int],4)

print ('European call based on BS simulation price is ' + str(np.round(bs_euro_call_price,4)),
       '\n  with 95% confidence level as : ', str(conf_interval))
```

```python
def euro_cal_price_bs_sim(NrSims=1000, T=1, nr_steps=1000, X0=100, r=0.1, sigma=0.2):
    
    nr_cases = len(NrSims)
    
    value_conf_interval = np.zeros((nr_cases, 4))
    
    for i in range(nr_cases):
        nr_paths = NrSims[i]
        Xs = sim_BS_Xt(T, nr_steps, X0, r, sigma, nr_paths)
        bs_euro_call_price = np.exp(-r * T) * np.mean(np.maximum(Xs[:,-1] - K,0))
        
        var_n = nr_paths /(nr_paths - 1) \
        * (np.mean(np.maximum(Xs[:,-1] - K,0) ** 2) - np.mean(np.maximum(Xs[:,-1] - K,0)) **2) 
        std_n = var_n ** 0.5
        
        half_int = 1.96 * std_n / nr_paths ** 0.5 
        conf_interval = np.round([bs_euro_call_price - half_int,  bs_euro_call_price + half_int],4)
        
        
        value_conf_interval[i, 0] = nr_paths
        value_conf_interval[i, 1] = bs_euro_call_price
        
        value_conf_interval[i, 2:4] = conf_interval
        
    return value_conf_interval    
    
```

```python
nr_sims = 10 ** np.arange(3,6)
value_conf_interval = euro_cal_price_bs_sim(NrSims=nr_sims)
```

```python

import pandas as pd
result = pd.DataFrame(value_conf_interval,
                      columns=['Nr_sims', 'numerical_price', 
                               '95% conf_lower_bound', '95% conf_upper_bound'])

print ('The therectical price is : 13.269676584660893')
result
```

#### Comment:
As we increase the sample size, the price is getting closer to the true price and the confidence interval narrows. 

```python
# simulate based on Euler methods
def sim_Euler(T, nr_step=1000, X0=100, r=0.1, sigma=0.2, nr_paths=1000):  
    #  We simulate with Euler scheme  

    dt = T / nr_step    
    Xs = np.zeros((nr_paths, 1))
    Xs[:] = X0  
    for i in range(nr_step): 
        Xs +=  r * Xs * dt + sigma * Xs * dt ** 0.5 * np.random.randn(nr_paths, 1)  
    return Xs 
```

```python
Euler_Xs = sim_Euler(T=1)
```

```python
# price European call based on the simualated path of Xt (simulation according to Euler scheme)
Euler_euro_call_price = np.exp(-r * T) * np.mean(np.maximum(Euler_Xs - K,0))

var_n = nr_paths /(nr_paths - 1) \
* (np.mean((np.exp(-r * T) * np.maximum(Euler_Xs - K,0)) ** 2) - np.mean(np.exp(-r * T) * np.maximum(Euler_Xs - K,0)) **2) 
std_n = var_n ** 0.5
    
std_n = var_n ** 0.5

# 95% interval corresponds to 1.96 * std_n / (sqrt(n))
half_int = 1.96 * std_n / nr_paths ** 0.5 
conf_interval = np.round([Euler_euro_call_price - half_int,  Euler_euro_call_price + half_int],4)

print ('European call based on Euler scheme price is ' + str(np.round(Euler_euro_call_price,4)),
       '\n  with 95% confidence level as : ', str(conf_interval))

```

```python
def euro_cal_price_Euler_sim(NrSims=10000, T=1, steps=1000, X0=100, r=0.1, sigma=0.2):
    
    nr_cases = len(steps)
    
    value_conf_interval = np.zeros((nr_cases, 5))
    
    nr_paths = NrSims
    for i in range(nr_cases):
        nr_step = steps[i]
        Xs = sim_Euler(T, nr_step, X0, r, sigma, nr_paths=nr_paths)
        Euler_euro_call_price = np.exp(-r * T) * np.mean(np.maximum(Xs - K,0))
        
        var_n = nr_paths /(nr_paths - 1) \
        * (np.mean((np.exp(-r * T) * np.maximum(Xs - K,0)) ** 2) - np.mean(np.exp(-r * T) * np.maximum(Xs - K,0)) **2) 
        std_n = var_n ** 0.5
        
        half_int = 1.96 * std_n / nr_paths ** 0.5 
        conf_interval = np.round([Euler_euro_call_price - half_int,  Euler_euro_call_price + half_int],4)
        
        
        value_conf_interval[i, 0] = nr_step
        value_conf_interval[i, 1] = Euler_euro_call_price
        
        value_conf_interval[i, 2] = np.abs(Euler_euro_call_price - 13.269676584660893)
        
        value_conf_interval[i, 3:5] = conf_interval
        
    return value_conf_interval 
```

```python
steps = [20, 30, 40, 50, 60, 70, 80, 90, 100, 200, 300, 400, 500,
        600, 700, 800, 900, 1000]
         #16000, 20000, 25000, 30000]
value_conf_interval = euro_cal_price_Euler_sim(steps=steps)
```

### Exercise 2: Asian call

```python
# simulate based on Euler methods
def sim_Eulerpath(T, nr_step=1000, X0=100, r=0.1, sigma=0.2, nr_paths=1000):  
    #  We simulate with Euler scheme  

    dt = T / nr_step    
    Xs = np.zeros((nr_paths, 1))
    Xs[:] = X0 
    
    AT = np.zeros((nr_paths, 1))
    AT = dt * X0 
    
    for i in range(nr_step): 
        Xs +=    r * Xs * dt + sigma * Xs * dt ** 0.5 * np.random.randn(nr_paths, 1)
        
        AT += dt * Xs
    return Xs, AT


```

```python
Xs, AT = sim_Eulerpath(T=1, nr_paths=10000, nr_step=2000)
Asian_price = np.mean(np.exp(-r * T) * np.maximum(AT-K, 0))
Asian_price
```

```python
nr_paths = 10000
var_n = nr_paths /(nr_paths - 1) \
    * (np.mean((np.exp(-r * T) * np.maximum(AT - K,0)) ** 2) - np.mean(np.exp(-r * T) * np.maximum(AT- K,0)) **2) 
    
std_n = var_n ** 0.5

# 95% interval corresponds to 1.96 * std_n / (sqrt(n))
half_int = 1.96 * std_n / nr_paths ** 0.5 

conf_interval = np.round([Asian_price - half_int,  Asian_price + half_int],4)

print ('Asian option based on Euler scheme price is ' + str(np.round(Asian_price,4)),
       '\n  with 95% confidence level as : ', str(conf_interval))
```

```python
def asian_sim(NrSims=1000, T=1, nr_steps=1000, X0=100, r=0.1, sigma=0.2, case='sim'):
       
    if case == 'sim': 
        nr_cases = len(NrSims)
    elif case == 'step': 
        nr_cases = len(nr_steps)
        
    
    value_conf_interval = np.zeros((nr_cases, 4))
    
    for i in range(nr_cases):
        
        if case == 'sim': 
            nr_paths = NrSims[i]
            nr_stepi = nr_steps
            nr = nr_paths
        elif case == 'step': 
            nr_stepi = nr_steps[i]
            nr_paths = NrSims
            nr = nr_stepi
            
        Xs, AT = sim_Eulerpath(T, nr_paths=nr_paths, nr_step=nr_stepi)
        
        Asian_price = np.mean(np.exp(-r * T) * np.maximum(AT-K, 0))
        
        var_n = nr_paths / (nr_paths - 1) * (np.mean((np.exp(-r * T) \
                        * np.maximum(AT - K,0)) ** 2) - np.mean(np.exp(-r * T) * np.maximum(AT- K,0)) **2) 
    
        std_n = var_n ** 0.5

        # 95% interval corresponds to 1.96 * std_n / (sqrt(n))
        half_int = 1.96 * std_n / nr_paths ** 0.5 

        conf_interval = np.round([Asian_price - half_int,  Asian_price + half_int],4)      
        value_conf_interval[i, 0] = nr
        value_conf_interval[i, 1] = Asian_price
        
        value_conf_interval[i, 2:4] = conf_interval
        
    return value_conf_interval  
```

```python
nr_sims = 10 ** np.arange(3,6)
value_conf_interval = asian_sim(NrSims=nr_sims)


import pandas as pd
result = pd.DataFrame(value_conf_interval,
                      columns=['Nr_sims', 'numerical_price', 
                               '95% conf_lower_bound', '95% conf_upper_bound'])

print ('The therectical price is : 13.269676584660893')
result
```

#### Comment: 
when the number of simulation increases, we see that the confidence interval narrows

```python
# simulate based on Euler methods
def sim_Asian_trapezoidal(T=1, nr_step=1000, X0=100, r=0.1, sigma=0.2, nr_paths=1000):  
    #  We simulate with Euler scheme  

    dt = T / nr_step    
    Xtminus = np.zeros((nr_paths, 1))
    

    Xtminus[:] = X0 
    
    
    
    AT = np.zeros((nr_paths, 1))
   
    
    for i in range(nr_step): 
        Xtplus = Xtminus +  r * Xtminus * dt + sigma * Xtminus * dt ** 0.5 * np.random.randn(nr_paths, 1)
        
        AT += dt * (Xtplus + Xtminus) / 2
        Xtminus = Xtplus
        
    return AT


```

```python
def asian_sim_trapezoidal(NrSims=1000, T=1, nr_steps=1000, X0=100, r=0.1, sigma=0.2, case='sim'):
       
    if case == 'sim': 
        nr_cases = len(NrSims)
    elif case == 'step': 
        nr_cases = len(nr_steps)
        
    
    value_conf_interval = np.zeros((nr_cases, 5))
    
    for i in range(nr_cases):
        
        if case == 'sim': 
            nr_paths = NrSims[i]
            nr_stepi = nr_steps
            nr = nr_paths
        elif case == 'step': 
            nr_stepi = nr_steps[i]
            nr_paths = NrSims
            nr = nr_stepi
            
        AT = sim_Asian_trapezoidal(T, nr_paths=nr_paths, nr_step=nr_stepi)
        
        Asian_price = np.mean(np.exp(-r * T) * np.maximum(AT-K, 0))
        
        var_n = nr_paths / (nr_paths - 1) * (np.mean((np.exp(-r * T) \
                        * np.maximum(AT - K,0)) ** 2) - np.mean(np.exp(-r * T) * np.maximum(AT- K,0)) **2) 
    
        std_n = var_n ** 0.5

        # 95% interval corresponds to 1.96 * std_n / (sqrt(n))
        half_int = 1.96 * std_n / nr_paths ** 0.5 

        conf_interval = np.round([Asian_price - half_int,  Asian_price + half_int],4)      
        value_conf_interval[i, 0] = nr
        value_conf_interval[i, 1] = Asian_price
        
        value_conf_interval[i, 2:4] = conf_interval
        value_conf_interval[i, 4] = half_int * 2
        
    return value_conf_interval  
```

```python
nr_sims = [1000, 5000, 10000, 50000, 100000]


value_conf_interval = asian_sim_trapezoidal(NrSims=nr_sims)


import pandas as pd
result = pd.DataFrame(value_conf_interval,
                      columns=['Nr_sims', 'numerical_price', 
                               '95% conf_lower_bound', '95% conf_upper_bound','interval_length'])

result



```
